```python
roads = {
 'S1': ['S2', 'S3'],
 'S2': ['S1', 'S4'],
 'S3': ['S1', 'S5'],
 'S4': ['S2'],
 'S5': ['S3']
}
roads
```

```python
def find_reachable_shelters(roads , start):
  queue = [start]
  visited = set()
  reachable = []

  while queue:
    shelter = queue.pop(0)
    if shelter not in visited:
      reachable.append(shelter)
      visited.add(shelter)
      queue.extend(roads[shelter])

  return reachable

```
